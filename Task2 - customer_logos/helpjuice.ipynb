```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://helpjuice.com/case-studies"
headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }

r = requests.get(url, headers=headers)

soup = BeautifulSoup(r.content, "html.parser")
```

```python
containers = soup.find_all('div',class_='case-study')
```

```python
print(len(containers))
```

```python
containers
```

```python
# Extract image URLs with tag img and class "logo"
helpjuice_logo_urls = list(set([case_study.find('img', class_='logo')['src'] for case_study in containers]))

# Print the extracted URLs
for logo_url in helpjuice_logo_urls:
    print(logo_url)
```

```python
type(helpjuice_logo_urls)
```

```python
import os
# Extract file extensions
extensions = {os.path.splitext(link)[1].lower() for link in helpjuice_logo_urls}

# Print unique extensions
print("Unique extensions:", extensions)
```

```python
# Creating a folder named "helpjuice"
import os
from urllib.parse import urljoin

folder_name = "helpjuice_customers_logo"
os.makedirs(folder_name, exist_ok=True)
```

```python
for index, src in enumerate(helpjuice_logo_urls, start=1):
    # Getting the image content
    response = requests.get(src)

    # Extracting the image file name from the URL
    if src.endswith('.svg'):
        img_name = os.path.join(folder_name, f"customer{index}.svg")
    elif src.endswith('.png'):
        img_name = os.path.join(folder_name, f"customer{index}.png")
    elif src.endswith('.jpeg'):
        img_name = os.path.join(folder_name, f"customer{index}.jpeg")
    elif src.endswith('.jpg'):
        img_name = os.path.join(folder_name, f"customer{index}.jpg")
    elif src.endswith('.webp'):
        img_name = os.path.join(folder_name, f"customer{index}.webp")
    elif src.endswith('.gif'):
        img_name = os.path.join(folder_name, f"customer{index}.gif")
    elif src.endswith('.jfif'):
        img_name = os.path.join(folder_name, f"customer{index}.webp")
    else:
        # Handle other image formats if needed
        
        continue

    # Saving the image to the folder
    with open(img_name, 'wb') as img_file:
        img_file.write(response.content)

    print(f"Image saved: {img_name}")
```
